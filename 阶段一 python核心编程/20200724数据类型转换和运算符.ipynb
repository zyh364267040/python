# 数据类型转换函数

```python
# 1. float() -- 将函数转换成浮点型
num1 = 1
str1 = '10'
print(type(float(num1)))
print(float(num1))

print(float(str1))

# 2. str() -- 将数据转换成字符串型
print(type(str(num1)))

# 3. tuple() -- 将一个序列转换成元组
list1 = [10, 20, 30]
print(tuple(list1))

# 4. list() -- 讲一个序列转换成列表
t1 = (100, 200, 300)
print(list(t1))

# 5. eval() -- 计算字字符串中的有效python表达式，并返回一个对象
str2 = '1'
str3 = '1.1'
str4 = '(1000, 2000, 3000)'
str5 = '[1000, 2000, 3000]'
print(type(eval(str2)))
print(type(eval(str3)))
print(type(eval(str4)))
print(type(eval(str5)))
```

# 复合赋值运算符

```python
a = 10
a += 1
print(a)


b = 10
b -= 1
print(b)

# 注意：先算复合运算符右面的表达式，再算复合赋值运算
c = 10
c += 1 + 2
# c += 3 -- c = c + 3
print(c)

d = 10
d *= 1 + 2
print(d)
```

# 逻辑运算符

```python
a = 0
b = 1
c = 2

# 1. and: 与: 都真才真
print((a < b) and (c > b))
print((a > b) and (c > b))

# 2. or: 或: 一真则真，都假才假
print((a < b) or (c > b))
print((a > b) or (c > b))

# 3. not: 非: 取反
print(not False)
print(not (c < b))
```
