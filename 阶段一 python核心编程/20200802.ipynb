## 学员管理系统

```python
# 定义学院管理系统选项
def info_print():
    print('请选择功能：')
    print('1、添加学员')
    print('2、删除学员')
    print('3、修改学员信息')
    print('4、查询学员信息')
    print('5、显示所有学员信息')
    print('6、退出系统')
    print('-' * 20)

# 等待存储所有学员信息
info = []

# 定义添加学员函数
def info_add():
    """添加学员"""
    # 1. 用户输入学员学号、姓名、手机号
    new_id = input('请输入学号：')
    new_name = input('请输入姓名：')
    new_tel = input('请输入手机号：')
    # 2. 判断是否添加这个学员：如果学员姓名已经存在报错提示；如果姓名不存在添加数据
    global info
    # 2.1 不允许学号重复：判断用户输入的学号 和 列表里面字典的id对应的值 相等 提示
    for i in info:
        if new_id == i[id]:
            print('此学员学号已经存在')
            # return作用：退出当前函数，后面添加信息的代码不执行
            return
    
    # 2.2 如果输入的姓名不存在，添加数据；准备空字典，字典新增数据，列表追加字典
    info_dict = {}
    
    # 字典新增数据
    info_dict['id'] = int(new_id)
    info_dict['name'] = str(new_name)
    info_dict['tel'] = int(new_tel)
    
    # 列表追加数据
    info.append(info_dict)
    print(info)
    
# 定义删除学员函数
def info_del():
    """删除学员"""
    # 输入要删除的学员的学号
    del_id = input('请输入删除学员的学号：')
    global info
    # 判断该学员是否存在，如果存在则删除（删除字典）
    for i in info:
        if int(del_id) == i['id']:
            info.remove(i)
            print('该学员已删除')
            break
    else:
        print('该学员不存在')
    print(info)

# 定义修改学员函数
def info_modify():
    """修改学员信息"""
    # 输入要修改的学员学号
    modify_id = input('请输入要修改学员的学号')
    global info
    # 判断该学员是否存在，如果存在则对信息进行修改
    for i in info:
        if int(modify_id) == i['id']:
            j = int(input('请输入修改信息序号（修改name请输入1;修改tel请输入2）：'))
            if  j == 1:
                i['name'] = input('请输入修改后的name')
                print('该学员信息已修改完毕')
            elif j == 2:
                i['tel'] = int(input('请输入修改后的tel'))
                print('该学员信息已修改完毕')
            else:
                print('您输入序号有误')
            break
    else:
        print('该学员不存在')
    print(info)
    
# 定义查询学员信息函数
def info_search():
    """查询学员"""
    secrch_id = int(input('请输入学员的学号：'))
    global info
    for i in info:
        if secrch_id == i['id']:
            print('查询到的学员信息如下：')
            print(f"学员的学号是{i['id']},学员的姓名是{i['name']}学员的电话是{i['tel']}")
            break
    else:
        print('查无此人...')

# 定义显示所有学员信息函数
def info_all():
    """显示所有学员信息"""
    print("学号\t姓名\t手机号")
    for i in info:
        print(f"{i['id']}\t{i['name']}\t{i['tel']}")

# 系统循环调用，直到用户输入6，退出系统    
while True:
    info_print()
    
    # 用户选择功能序号
    user_num = eval(input('请输入功能选项序号：'))
    
    if user_num == 1:
        # print('添加')
        info_add()
    elif user_num == 2:
        # print('删除')
        info_del()
    elif user_num == 3:
        # print('修改')
        info_modify()
    elif user_num == 4:
        # print('查询')
        info_search()
    elif user_num == 5:
        # print('显示所有')
        info_all()
    elif user_num == 6:
        # print('退出系统')
        exit_flag = input("确定退出系统吗？yes or no")
        if exit_flag == "yes":
            print("退出系统")
            break
    else:
        print('您输入的功能序号有误，请重新输入')
```
