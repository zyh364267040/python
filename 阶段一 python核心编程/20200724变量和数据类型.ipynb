# 1.1 格式化符号


格式化字符串除了%s，还可以写为 f'{表达式}'

# 格式化练习

```python
age = 18
name = 'zhou'
weight = 74.5
stu_id = 2010123869
stu_id2 = 1


# 1. 今年我的年龄是x岁
print('今年我的年龄是%d岁' % age)

# 2. 我的名字是x
print('我的名字是%s' % name)

# 3. 我的体重是x公斤
print('我的体重是%.2f公斤' % weight)
# %.2f 表示小数点后面保留两位小数

# 4. 我的学号是x
print('我的学号是%d' % stu_id2)

# 4.1 我的学号是001
print('我的学号是%03d' % stu_id2)
print('我的学号是%03d' % stu_id)
print("-" * 5)
# 使用f格式化打印学号001
print(f"我的学号是{stu_id:03d}")
print(f"我的学号是{stu_id2:03d}")
# %03d 表示不够3位数，用0在数字前面补齐，若超出3位数，则按原数输出

# 5. 我的名字是x，今年x岁了
print('我的名字是%s，今年%d岁了' % (name, age))
# 5.1 我的名字是x，明年x岁了
print('我的名字是%s，明年%d岁了' % (name, age + 1))

# 6. 我的名字是x，今年x岁了，体重x公斤，学号是x
print('我的名字是%s，今年%d岁了，体重%.2d公斤，学号是%03d' % (name, age, weight, stu_id2))
```

# 格式化字符串扩展

```python
name = 'zhou'
age = 18
weight = 74.5


# 我的名字是x，今年x岁了，体重x公斤
print('我的名字是%s，今年%s岁了，体重%s公斤' % (name, age, weight))
```

# f格式化字符串

```python
name = 'zhou'
age = 18

# 我的名字是x，今年x岁了
print('我的名字是%s，今年%s岁了' % (name, age))

# 语法f'{表达式}'
print(f'我的名字是{name}，今年{age}岁了')
```

# 1.3 转义字符
\n ：换⾏。

\t ：制表符，⼀个tab键（4个空格）的距离。

# 转义字符

```python
print('hello')
print('world')

print('hello\nworld')

print('\tabcd')

# \n 换行字符
# \t 空4个空格，一个tab
```

# print的结束符
