```python
age = 18
name = 'zhou'
weight = 74.5
stu_id = 2010123869
stu_id2 = 1


# 1. x
print('%d' % age)

# 2. x
print('%s' % name)

# 3. x
print('%.2f' % weight)
# %.2f 

# 4. x
print('%d' % stu_id2)

# 4.1 001
print('%03d' % stu_id2)
print('%03d' % stu_id)
print("-" * 5)
# f001
print(f"{stu_id:03d}")
print(f"{stu_id2:03d}")
# %03d 303

# 5. xx
print('%s%d' % (name, age))
# 5.1 xx
print('%s%d' % (name, age + 1))

# 6. xxxx
print('%s%d%.2d%03d' % (name, age, weight, stu_id2))
```
