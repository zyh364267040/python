```python
def info_print():
    """系统功能函数"""
    print("----------系统功能如下----------")
    print("1.添加学生")
    print("2.删除学生")
    print("3.修改学生")
    print("4.查询学生")
    print("5.显示所有学生")
    print("6.退出系统")
    print("7.保存信息")

    
# 学生信息存储列表
student_list = []


def all_student():
    """显示所有学生"""
    print("学号\t姓名\t手机号")
    for i in student_list:
        print(f"{i['id']}\t{i['name']}\t{i['tel']}")
        
    
def add_student():
    """添加学生"""
    add_id = input("请输入要添加的学生的学号:")
    # 如果输入的id学生列表已存在，显示该学号学生已存在，并退出该函数
    # 全局变量
    global student_list
    for i in student_list:
        if add_id == i["id"]:
            print(f"学号:{add_id}学生已存在.")
            return
        
    add_name = input("请输入要添加的学生的姓名:")
    add_tel = input("请输入要添加的学生的手机号:")
    student_dict = {}
    student_dict["id"] = add_id
    student_dict["name"] = add_name
    student_dict["tel"] = add_tel
    student_list.append(student_dict)
    
    print("添加后系统内的学生为:")
    all_student()
    
    
def del_student():
    """删除学生"""
    del_id = input("请输入删除学生的学号:")
    global student_list
    for i in student_list:
        if i["id"] == del_id:
            student_list.remove(i)
            print(f"学号:{del_id}学生已删除.")
            print("删除后系统内的学生为:")
            all_student()
    else:
        print(f"未找到学号:{del_id}学生")


def modify_student():
    """修改学生"""
    modify_id = input("请输入要修改的学生的学号:")
    global student_list
    for i in student_list:
        if modify_id == i["id"]:
            i["name"] = input(f"请输入学号:{modify_id}学生修改后的名字:")
            i["tel"] = input(f"请输入学号:{modify_id}学生修改后的手机号:")
            print("修改后系统内的学生信息:")
            all_student()
            break
    else:
        print(f"未找到学号:{modify_id}的学生.")


def search_student():
    """查询学生"""
    search_id = input("请输入要查询的学生的id:")
    for i in student_list:
        if search_id == i["id"]:
            print("您要查询的学生信息如下:")
            print("学号\t姓名\t手机号")
            print(f"{i['id']}\t{i['name']}\t{i['tel']}")
            break
    else:
        print(f"未找到学号为{search_id}的学生.")
        

# 保存学生信息
def save_student():
    """保存学生信息"""
    # 1.打开文件
    student_f = open("student.txt", "w", encoding="utf-8")
    # 2.把全局变量学生信息保存到文件 -- 写入到文件，文件内数据都是字符串
    student_f.write(str(student_list))
    # 3.关闭文件
    student.close()
    
    
# 加载数据 -- 加载文件中的数据
# 如果student.txt不存在，说明没有数据需要加载，也就是说明打开文件等操作不需要执行
def load_student():
    import os
    # os.path.exists(目标文件) 如果文件存在返回True，否则返回False
    if os.path.exists("student.txt"):
        # 1.打开
        student_f = open("student.txt", "r", encoding="utf-8")
        
        # 2.读取文件数据
        con = student_f.read()
        # con 保存全局列 list 要保存数据，修改global
        list1 = eval(con)
        global student_list
        student_list = list1
        
        # 3.关闭
        student_f.close()
        

# 加载数据的函数调用放在循环的外面，进入系统只加载一次数据，数据就保存在全局变量列表(内存)
load_student()


# def all_student():
#     """显示所有学生"""
#     print("学号\t姓名\t手机号")
#     for i in student_list:
#         print(f"{i['id']}\t{i['name']}\t{i['tel']}")

    
while True:
    info_print()
    
    
    # 输入序号对应的功能
    num = eval(input("请输入需要的功能序号:"))
    if num == 1:
        # 添加学生
        add_student()
    elif num == 2:
        # 删除学生
        del_student()
    elif num == 3:
        # 修改学生
        modify_student()
    elif num == 4:
        # 查询学生
        search_student()
    elif num == 5:
        # 显示所有学生
        print("系统内的学生为:")
        all_student()
    elif num == 6:
        print("退出系统")
        break
        # 退出系统
    elif num == 7:
        print("保存信息")
        save_student()
    else:
        print("输入内容不正确，请重新输入")
```
