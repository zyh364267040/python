# Linux命令一

### 查看目录命令
- ls：查看当前目录信息
    - -l：以列表方式显示
    - -h：文件大小单位显示，默认是字节
    - -a：显示隐藏文件和隐藏目录
    - ll 相当于 ls -la
    
    
- tree：以树状方式显示目录信息
    - tree 文件夹名：表示显示该文件夹内目录信息
- pwd：查看当前目录路径
- clear：清除终端内容

### 终端字体控制
- ctr + alt + t:快速启动终端
- ctr + shift + "+":表示终端字体放大
- ctr + "-":表示终端字体缩小

### 切换目录命令
- cd目录：切换到指定目录
- cd ~：切换到当前用户的主目录
- cd ..：切换到上一级目录
    - cd ../..：切换到上一级的上一级目录
- cd .：切换到当前目录
- cd -：切换到上一次目录
- 注意：
    - cd命令切换目录时，这个目录必须存在
    - cd后面不写目录等价于cd~

### 绝对路径和相对路径
- 绝对路径：从根目录算起的路径
- 相对路径：从当前目录算起的路径
    - cd ../Desktop：表示回到上一级目录的Desktop目录中
- 绝对路径和相对路径是路径的两种表现形式，应遵循最短路径原则（使用最短的路径操作某个文件或者目录）
    - 如果切换的目录离当前目录近可以使用相对路径
    - 如果切换的目录离根目录近可以使用绝对路径
    - 如果切换的目录离根目录和当前目录都一样那么使用绝对路径，因为这样路径很清楚和明确，方便阅读
- 自动补全
    - 当敲出 文件或者目录或者命令 的前几个字母之后，按下tab键，如果输入的没有歧义系统会自动补全
    - 当切换目录或者使用命令的时候，如果不确定当前目录下有哪些子目录或者其他命令，可以通过两次tab键查看

### 创建、删除文件及目录命令
- touch 文件名：创建指定文件
    - touch 1.txt 2.txt 3.txt：创建多个文件
- mkdir 目录名：创建目录（文件夹）
    - mkdir AA BB CC：创建多个文件夹
    - -p：创建所依赖的文件夹
    - mkdir A/B 如果A不存在会报错
    - mkdir A/B -p A不存在加-p就不会报错
- 创建隐藏文件，在文件/目录前面加"."
- rm 文件名或者目录名：删除指定文件或者目录
    - rm 文件名 -r：删除非空目录
        - -r表示递归删除目录及其内容
    - rm命令选项
    - -i：交互式提示
    - -r：递归删除目录及其内容
    - -f：强制删除，忽略不存在的文件，无需提示
    - -d：删除空目录
- 删除空目录使用 "rmdir 目录名" 或者 "rm -d 目录名"
- 删除非空目录使用 "rm -r 目录名"

### 复制、移动问价及目录命令
- cp：拷贝文件、拷贝目录
    - 拷贝目录需要加上-r，-r表示递归拷贝目录及其内容
    - -i：交互式提示(提示是否覆盖)
    - -r：递归拷贝目录及其内容
    - -v：显示拷贝后的路径描述
    - -a：保留文件的原有权限
        - 主要针对其他用户
        
        - 还可以拷贝文件夹，并且文件夹里面的文件权限也不丢失
        
- mv：移动文件、移动目录、重命名
    - mv移动目录的时候，移动到另一个目录，如果另一目录存在，则移到另一个目录中；若该目录不存在，则将该目录重命名为另一目录
    - -i：交互式提示(提示是否覆盖)
    - -v：显示移动后的路径描述

### 终端命令格式说明
- **command** [-options][parameter]
    - command：命令名
    - [-options]：选项，可以有0个、1个或多个选项，多个选项可以合并
    - [parameter]：参数，可以有0个、1个或多个参数
    - []：代表可选
- 对于命令，他的选项和参数一般情况下没有顺序要求，但是scp命令是必须先选项然后在跟上参数

### 查看命令帮助
- --help 使用说明：命令 --help
- man 使用说明：man 命令
    - 空格：显示下一屏信息
    - 回车：显示下一行信息
    - b：显示上一屏信息
    - f：显示下一屏信息
    - q：退出

### 重定向命令
- 也称为输出重定向，把在终端执行命令的结果保存到目标文件
- ">"：如果文件存在会覆盖原有文件内容，相当于物件操作中的"w"模式，每次写入最新的数据，原有数据不保留
- ">>"：如果文件存在会追加写入文件末尾，相当于文件操作中的"a"模式，每次在原有数据的基础上进行追加，原有数据会保留
- 只要在终端能显示信息的命令都可以使用重定向
